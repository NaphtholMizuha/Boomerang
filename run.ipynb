```python
from FedDPR.config import toml2cfg
from FedDPR.algorithm import fetch_algorithm

cfg = toml2cfg("config/default.toml")

cfg.algorithm = 'feddpr'
cfg.db.table = 'feddpr'
cfg.n_rounds = 100

alg = fetch_algorithm(cfg)
alg.run()
```

```python
import networkx as nx
import matplotlib.pyplot as plt


# 完全图 (Complete Graph)
def draw_complete_graph(n):
    G = nx.complete_graph(n)
    plt.figure(figsize=(5, 5))
    nx.draw(G, with_labels=True, node_color='skyblue', font_weight='bold', node_size=500, font_size=12)
    plt.show()

# 二部图 (Bipartite Graph)
def draw_bipartite_graph(m, n):
    G = nx.bipartite.random_graph(m, n, 1)
    plt.figure(figsize=(5, 5))
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightgreen', font_weight='bold', node_size=500, font_size=12)
    plt.show()

# 星状图 (Star Graph)
def draw_star_graph(n):
    G = nx.star_graph(n-1)  # n-1 because star_graph includes center node
    plt.figure(figsize=(5, 5))
    nx.draw(G, with_labels=True, node_color='orange', font_weight='bold', node_size=500, font_size=12)
    plt.show()

# 环状图 (Cycle Graph)
def draw_cycle_graph(n):
    G = nx.cycle_graph(n)
    plt.figure(figsize=(5, 5))
    nx.draw(G, with_labels=True, node_color='lightcoral', font_weight='bold', node_size=500, font_size=12)
    plt.show()

# 绘制图形
draw_complete_graph(6)   # 完全图 K_4
draw_bipartite_graph(2,4)  # 二部图 K_3,4
draw_star_graph(8)  # 星状图 S_5
draw_cycle_graph(9)  # 环状图 C_4
```
